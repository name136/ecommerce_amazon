Your dataset contains order_date in multiple formats: 'DD/MM/YYYY', 'DD-MM-YY', 'YYYY-MM-DD', and some invalid entries like '32/13/2020'. Clean and standardize all dates to 'YYYY-MM-DD' format, handling invalid dates appropriately.

```python
import pandas as pd
import re

years = range(2015, 2026)  # Covers df_2015 to df_2025

# Store rows with bad or inconsistent date formats
bad_dates_report = {}

# Helper: Try parsing date
def try_parse_date(val):
    try:
        pd.to_datetime(val)
        return True
    except:
        return False

# Helper: Guess date format using regex
def detect_date_format(value):
    value = str(value)
    if re.match(r'\d{4}-\d{2}-\d{2}', value):
        return 'YYYY-MM-DD'
    elif re.match(r'\d{2}/\d{2}/\d{4}', value):
        return 'MM/DD/YYYY'
    elif re.match(r'\d{4}/\d{2}/\d{2}', value):
        return 'YYYY/MM/DD'
    elif re.match(r'[A-Za-z]{3,9} \d{1,2}, \d{4}', value):
        return 'Month Day, Year'
    else:
        return 'Unknown'

# Loop through all DataFrames
for year in years:
    df = globals().get(f'df_{year}')
    if df is None:
        print(f"❌ df_{year} not found.")
        continue

    print(f"\n🔍 Checking df_{year} for 'order_date'...")

    if 'order_date' not in df.columns:
        print("  ⚠️ 'order_date' column not found.")
        continue

    # Get unique non-null date values
    date_values = df['order_date'].dropna().astype(str).unique()

    bad_entries = []

    for val in date_values:
        if not try_parse_date(val):
            fmt = detect_date_format(val)
            bad_entries.append((val, fmt))

    if bad_entries:
        bad_dates_report[year] = bad_entries
        print(f"  ⚠️ Found {len(bad_entries)} problematic date values.")
    else:
        print("  ✅ All 'order_date' values are valid and parseable.")

# === Summary Report ===
print("\n\n📦 Summary of unrecognized/inconsistent 'order_date' formats:\n")

for year, entries in bad_dates_report.items():
    print(f"🗓️ df_{year}:")
    for val, fmt in entries:
        print(f"  Value: {val} | Detected Format: {fmt}")
```

Customer ratings appear in various formats: '5.0', '4 stars', '3/5', '2.5/5.0', and some missing values. Standardize all ratings to numeric scale 1.0-5.0, handling inconsistent formats and missing values strategic

```python
import pandas as pd
import os

input_dir = '/content/D:/course/cleaned_price_files'

for year in range(2015, 2026):
    file_path = os.path.join(input_dir, f'price_clean_df_{year}.csv')

    if not os.path.exists(file_path):
        print(f"❌ File not found: {file_path}")
        continue

    df = pd.read_csv(file_path)

    if 'customer_rating' in df.columns:
        dtype = df['customer_rating'].map(type).value_counts()
        print(f"\n📁 df_{year} — `customer_rating` data types:")
        print(dtype)
    else:
        print(f"⚠️ 'customer_rating' column not found in df_{year}")
```

```python
import pandas as pd
import numpy as np
import re

def clean_customer_rating(col):
    def parse_rating(value):
        if pd.isna(value):
            return np.nan

        value = str(value).strip().lower()

        # Case: '4 stars', '2 star'
        if 'star' in value:
            match = re.search(r'(\d+(\.\d+)?)', value)
            if match:
                num = float(match.group(1))
                if 1 <= num <= 5:
                    return round(num, 2)

        # Case: '3/5', '2.5/5.0' etc.
        if '/' in value:
            parts = re.split(r'\s*/\s*', value)
            if len(parts) == 2:
                try:
                    num = float(parts[0])
                    denom = float(parts[1])
                    if denom != 0:
                        scaled = (num / denom) * 5
                        if 0 <= scaled <= 5:
                            return round(scaled, 2)
                except:
                    return np.nan

        # Plain numeric string: '5.0', '4'
        if re.match(r'^\d+(\.\d+)?$', value):
            num = float(value)
            if 0 <= num <= 5:
                return round(num, 2)

        # Anything else: NaN
        return np.nan

    return col.apply(parse_rating)

def clean_ratings_in_df(df, col_name='customer_rating', output_col='customer_rating_clean'):
    if col_name not in df.columns:
        raise ValueError(f"Column '{col_name}' not found in DataFrame")

    df[output_col] = clean_customer_rating(df[col_name])
    return df

# Example usage:
# df = clean_ratings_in_df(df, col_name='customer_rating')
```

```python
# To handle NaN, we extract product id with NaN. check their rating in other rows

import pandas as pd
import numpy as np
import os

# Correct path where cleaned rating files are stored
data_path = '/content/D:/course/rating_cleaned_files'
years = range(2015, 2026)

for year in years:
    print(f"\n🔍 Processing year: {year}")

    rating_clean_file = os.path.join(data_path, f'rating_clean_df_{year}.csv')

    if not os.path.exists(rating_clean_file):
        print(f"❌ File not found: {rating_clean_file}")
        continue

    df = pd.read_csv(rating_clean_file)

    if 'product_id' not in df.columns or 'customer_rating_clean' not in df.columns:
        print(f"⚠️ Required columns missing in df_{year}")
        continue

    # Get rows with missing ratings
    missing_ratings = df[df['customer_rating_clean'].isna()]
    total_missing = len(missing_ratings)

    # Count how many product_ids with NaN have ratings elsewhere
    recovered_count = 0

    for pid in missing_ratings['product_id'].dropna().unique():
        same_product_ratings = df[
            (df['product_id'] == pid) &
            (df['customer_rating_clean'].notna())
        ]
        if not same_product_ratings.empty:
            recovered_count += 1

    # Summary
    percent_recoverable = (recovered_count / total_missing) * 100 if total_missing else 0
    print(f"📝 Total missing ratings: {total_missing}")
    print(f"🔁 Recoverable ratings using same product_id: {recovered_count} ({percent_recoverable:.2f}%)")


```

```python
import os
import pandas as pd
import numpy as np

# Set your data path
data_path = 'D:/course/rating_cleaned_files'

# Cleaning function for delivery_days
def clean_delivery_day(val):
    if pd.isna(val):
        return np.nan

    val_str = str(val).strip().lower()

    if val_str == 'same day':
        return 0
    elif val_str == 'express':
        return 1
    elif '-' in val_str:
        try:
            # Handle values like '1-2 days'
            nums = [int(s) for s in val_str.replace('days', '').split('-') if s.strip().isdigit()]
            if len(nums) == 2:
                return sum(nums) / 2
        except:
            return np.nan
    else:
        try:
            num = float(val_str)
            if num < 0 or num > 30:
                return np.nan
            return num
        except:
            return np.nan

# Loop through all files and clean the delivery_days column
for year in range(2015, 2026):
    file_name = f'city_clean_df_{year}.csv'
    file_path = os.path.join(data_path, file_name)

    if os.path.exists(file_path):
        df = pd.read_csv(file_path)

        if 'delivery_days' in df.columns:
            df['delivery_days'] = df['delivery_days'].apply(clean_delivery_day)
            df.to_csv(file_path, index=False)
            print(f'✅ Cleaned delivery_days in file for year {year}')
        else:
            print(f'⚠️ No "delivery_days" column in file for year {year}')
    else:
        print(f'❌ File not found: {file_name}')
```

Question 8: Identify and handle duplicate transactions where the same customer, product, date, and amount appear multiple times. Some duplicates are genuine (bulk orders) while others are data errors. Develop a strategy to distinguish and handle both cases.

So the following strategy was used for differentiating duplicates from bulk order:

All fields including order_id are identical	❌ Drop (duplicate)
Same customer/product/date, 2–3 exact repeats (same amount)	✅ Keep (bulk order)
Same combo, but final_amount_inr varies	✅ Keep all
Same combo 4+ times, all fields same incl. amount	⚠️ Flag for review
quantity column available and value >1	✅ Keep (bulk)
order_date is missing but other fields match	✅ Keep

```python
# remove duplicates

# Group by relevant fields excluding amount
group_cols = ['customer_id', 'product_id', 'product_name', 'subtotal_inr', 'order_date']

# Count how many times each group appears
df['dup_group_count'] = df.groupby(group_cols)['customer_id'].transform('count')

# Count how many unique final_amounts per group
df['unique_amounts'] = df.groupby(group_cols)['final_amount_inr'].transform('nunique')

# Classification function
def classify(row):
    if pd.isna(row['order_date']):
        return 'Keep'
    elif row['unique_amounts'] > 1:
        return 'Keep'  # Pricing varied — keep all entries
    elif 'quantity' in df.columns and not pd.isna(row['quantity']) and row['quantity'] > 1:
        return 'BulkOrder'
    elif row['dup_group_count'] <= 3:
        return 'BulkOrder'
    else:
        return 'SuspectedDuplicate'

df['duplicate_status'] = df.apply(classify, axis=1)

```

```python
import os
import pandas as pd

# Folder path
data_path = 'D:/course/rating_cleaned_files'

# Base columns for group comparison (excluding amount!)
group_cols = ['customer_id', 'product_id', 'product_name', 'subtotal_inr', 'order_date']

for year in range(2015, 2026):
    file_name = f'city_clean_df_{year}.csv'
    file_path = os.path.join(data_path, file_name)

    if not os.path.exists(file_path):
        print(f'❌ File not found: {file_name}')
        continue

    df = pd.read_csv(file_path)
    original_len = len(df)

    # Drop fully identical rows (including order_id if present)
    drop_cols = group_cols + ['final_amount_inr']
    if 'order_id' in df.columns:
        drop_cols.append('order_id')
    df_deduped = df.drop_duplicates(subset=drop_cols, keep='first')
    after_exact_drop = len(df_deduped)

    # Add duplicate group count (excluding final_amount_inr)
    df_deduped['dup_group_count'] = df_deduped.groupby(group_cols)['customer_id'].transform('count')

    # Count distinct final_amount_inr values within each group
    df_deduped['unique_amounts'] = df_deduped.groupby(group_cols)['final_amount_inr'].transform('nunique')

    # Classify entries
    def classify(row):
        if pd.isna(row['order_date']):
            return 'Keep'
        elif row['unique_amounts'] > 1:
            return 'Keep'
        elif 'quantity' in df_deduped.columns and not pd.isna(row['quantity']) and row['quantity'] > 1:
            return 'BulkOrder'
        elif row['dup_group_count'] <= 3:
            return 'BulkOrder'
        else:
            return 'SuspectedDuplicate'

    df_deduped['duplicate_status'] = df_deduped.apply(classify, axis=1)

    # Separate suspected duplicates
    suspected = df_deduped[df_deduped['duplicate_status'] == 'SuspectedDuplicate']
    cleaned_df = df_deduped[df_deduped['duplicate_status'] != 'SuspectedDuplicate']
    cleaned_len = len(cleaned_df)

    # Save cleaned file
    cleaned_df.to_csv(file_path, index=False)

    # Optionally: export suspected duplicates for audit
    if not suspected.empty:
        review_path = file_path.replace('.csv', '_suspected_duplicates.csv')
        suspected.to_csv(review_path, index=False)

    # Report
    removed_count = original_len - cleaned_len
    print(f"📁 {file_name}")
    print(f"   🔸 Original rows: {original_len}")
    print(f"   🔸 After dropping exact duplicates: {after_exact_drop}")
    print(f"   🔸 Suspected duplicates removed: {removed_count}")
    print(f"   ✅ Cleaned file saved.")
    if not suspected.empty:
        print(f"   📤 Suspected duplicates exported to: {os.path.basename(review_path)}")
```

Proposed work flow

Loop this for all the years.

3. Plot a line graph to check the trend and growth






```python
# Count unique values and their frequencies
return_status_counts = all_df_clean['return_status'].value_counts(dropna=False)

print(return_status_counts)
```

```python
# Count NaN values
nan_count = all_df_clean['return_status'].isna().sum()

# Count blank strings (assuming empty string means blank)
blank_count = (all_df_clean['return_status'].astype(str).str.strip() == '').sum()

print(f"NaN count: {nan_count}")
print(f"Blank count: {blank_count}")
```

```python
### REVENUE GENERATED YEAR WISE  ####

# Ensure year column exists and return_status is lowercase for matching
all_df_clean['return_status_lower'] = all_df_clean['return_status'].str.lower()

# Filter rows where return_status is 'delivered'
delivered_df = all_df_clean[all_df_clean['return_status_lower'] == 'delivered']

# Group by year and sum final_amount_inr
yearly_sum = delivered_df.groupby('year')['final_amount_inr'].sum().sort_index()

print(yearly_sum)
```

```python
import matplotlib.pyplot as plt
import numpy as np

years = yearly_sum.index.values
revenue = yearly_sum.values

# Fit linear trend line
coefficients = np.polyfit(years, revenue, 1)
trendline = np.poly1d(coefficients)

plt.scatter(years, revenue, color='blue', label='Yearly Revenue')
plt.plot(years, trendline(years), color='red', label='Linear Trend Line')

plt.xlabel('Year')
plt.ylabel('Revenue (INR)')
plt.title('Yearly Revenue with Linear Trend Line')
plt.legend()
plt.show()


```
